### **Use `get()` for quick lookups**

The `get()` method is designed for simple, equality-based queries on a single model.
It is ideal for fast inspections, previews, and exploratory work when you just need to retrieve records by matching column values (e.g., symbol, ID, or code) without writing a full SQLAlchemy statement.

Typical use cases:

* Look up a gene, variant, or disease by identifier
* Quickly inspect records in a table
* Run simple filters using `column = value`
* Keep notebook code short and readable

Example:

```python
from biofilter import Biofilter
bf = Biofilter()
```

```python
df = bf.query.get("GeneMaster", symbol="A1BG")
df
```

### **Use `run_model()` for full SQLAlchemy queries**

The `run_model()` method executes explicit SQLAlchemy statements and offers full control over the query logic.
It should be used whenever you need joins, multiple conditions, `IN` clauses, ordering, aggregation, or complex filtering across tables.

Typical use cases:

* Joining multiple models
* Filtering with `IN`, ranges, or custom expressions
* Selecting specific columns
* Aggregations (`count`, `group_by`, etc.)
* Reproducible and structured queries for reports

Example:

```python
stmt = (
    bf.query.select(bf.query.GeneMaster)
    .where(bf.query.GeneMaster.symbol == "A1BG")
    .limit(10)
)

df = bf.query.run_model(stmt)
df
```

### Summary

* **`get()`** â†’ simple, fast, equality-based lookups

A good rule of thumb:

> *Start with `get()` for exploration. Switch to `run_model()` when your query becomes more complex.*
