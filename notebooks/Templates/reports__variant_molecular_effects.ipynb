<h1> 📘 Biofilter — Report: Variant Molecular Effects </h1>

This notebook demonstrates how to use the `variant_molecular_effects` report to retrieve:
- variants overlapping genomic regions,
- molecular consequences from VEP-like annotations,
- resolved consequence/impact/biotype labels.

### Methods used
- `bf.report.explain("variant_molecular_effects")`
- `bf.report.available_columns("variant_molecular_effects")`
- `bf.report.example_input("variant_molecular_effects")`
- `bf.report.run("variant_molecular_effects", **params)`


--------


### 1. Start Biofilter


```python
from biofilter import Biofilter

# Uses db_uri from .biofilter.toml if available
bf = Biofilter(debug_mode=False)
```

------


### 2. Inspect the report contract


```python
bf.report.explain("variant_molecular_effects")
```

```python
bf.report.available_columns("variant_molecular_effects")
```

```python
bf.report.example_input("variant_molecular_effects")
```

---------


### 3. Run with built-in example input


```python
import time

start = time.time()
df = bf.report.run_example("variant_molecular_effects")
elapsed = time.time() - start

print(f"Rows: {len(df)} | elapsed: {elapsed:.2f}s")
df.head()
```

--------


### 4. Run with custom region inputs (string + dict)


```python
custom_items = [
    "chrY:2781644:2781644",
    "chr1:55516888:55516888",
    {"chromosome": "7", "start": 55019017, "end": 55019017},
]

df_custom = bf.report.run(
    "variant_molecular_effects",
    items=custom_items,
    range_up=50,
    range_down=50,
    emit_not_found_rows=True,
    include_variant_only_rows=True,
    limit_variants_per_input=500,
)

print(f"Rows: {len(df_custom)}")
df_custom.head()
```

```python
# Quick status breakdown for demo
if not df_custom.empty and "Status" in df_custom.columns:
    df_custom["Status"].value_counts(dropna=False)
```

--------


### 5. Run from input file (one region per line)


```python
from pathlib import Path

tmp_file = Path("/tmp/variant_regions_demo.txt")
tmp_file.write_text("".join([
    "chr1:55516888:55516888",
    "chr7:55019017:55019017",
    "chrX:2781644:2781644",
    ])
)

df_file = bf.report.run(
    "variant_molecular_effects",
    input_path=str(tmp_file),
    range_up=0,
    range_down=0,
    effect_query_chunk_size=2000,
)

print(f"Rows: {len(df_file)}")
df_file.head()
```

--------


### 6. Suggested view for presentation


```python
display_cols = [
    "Input",
    "Status",
    "Variant Key",
    "Chromosome",
    "Position Start",
    "Gene Symbol",
    "Consequence",
    "Impact",
    "Biotype",
    "LoF",
]

present_df = df_custom[[c for c in display_cols if c in df_custom.columns]].copy()
present_df.head(20)
```
