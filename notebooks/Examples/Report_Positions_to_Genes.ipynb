# üß¨ Example Report: Positions to Genes
This notebook demonstrates how to run the Position ‚Üí Genes report using Biofilter3R.

```python
from biofilter import Biofilter
```

```python
# Instance of Biofilter
bf = Biofilter("postgresql+psycopg2://bioadmin:bioadmin@localhost/biofilter")
```

```python
# Create an input list
position_list = [
    ("Y", 19568371),
    ("Y", 9999999999),]
```

```python
# Run Report
result = bf.report.run_report(
    "report_position_to_gene",
    assembly='GRC38',
    input_data=position_list,)
```
