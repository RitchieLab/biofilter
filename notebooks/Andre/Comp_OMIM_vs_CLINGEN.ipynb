```python
import pandas as pd
```

```python
# Load OMIM raw file (skip comment lines)
file_path = "omim.txt"
df_omim = pd.read_csv(
    file_path,
    sep="\t",
    comment="#",  # jump headed and copyright
    names=[
        "mim_number",
        "entry_type",
        "entrez_id",
        "hgnc_symbol",
        "ensembl_id",
    ],
    dtype=str
)

# Clean data: remove linhas completamente vazias
df_omim = df_omim.dropna(how="all")

# Reset index
df_omim.reset_index(drop=True, inplace=True)

print(df_omim.head(20))
print(df_omim.dtypes)
```
