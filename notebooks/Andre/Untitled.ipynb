```python
from biofilter import Biofilter
import biofilter.db.models as models
import pandas as pd
```

```python
db_uri = "postgresql+psycopg2://bioadmin:bioadmin@localhost/biofilter"
bf = Biofilter(db_uri)
```

```python
s = bf.db.get_session()
```

```python
genes = s.query(models.GeneMaster).limit(20).all()
for g in genes:
    print(g.symbol, " - ", g.hgnc_status ," - ", g.chromosome)
```

```python
results = (
    s.query(
        models.ETLProcess.id.label("process_id"),
        models.DataSource.name.label("data_source_name"),
        models.DataSource.data_type,
        models.ETLProcess.global_status,
        models.ETLProcess.extract_start,
        models.ETLProcess.extract_end,
        models.ETLProcess.extract_status,
        models.ETLProcess.transform_start,
        models.ETLProcess.transform_end,
        models.ETLProcess.transform_status,
        models.ETLProcess.load_start,
        models.ETLProcess.load_end,
        models.ETLProcess.load_status,
    )
    .join(models.DataSource, models.ETLProcess.data_source_id == models.DataSource.id)
    .order_by(models.ETLProcess.id.desc())
    .all()
)

data = [row._asdict() for row in results]

df = pd.DataFrame(data)
df
```
